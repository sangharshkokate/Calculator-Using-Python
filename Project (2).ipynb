## write a program to build a simple calulator :

```python
# Simple Calculator Program

print("Welcome to Sangharsh Calculator:")

name = input("Please Enter Your Name: ")
print(f"Hello {name}! Let's calculate something.\n")

try:
    num1 = float(input("Enter your first number: "))
    num2 = float(input("Enter your second number: "))

    print("\nSelect your operation:")
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")

    operation = input("Select your operation (1/2/3/4): ")

    if operation == '1':
        result = num1 + num2
        print(f"\nResult: {num1} + {num2} = {result}")

    elif operation == '2':
        result = num1 - num2
        print(f"\nResult: {num1} - {num2} = {result}")

    elif operation == '3':
        result = num1 * num2
        print(f"\nResult: {num1} × {num2} = {result}")

    elif operation == '4':
        if num2 == 0:
            print("\nError: Division by zero is not allowed.")
        else:
            result = num1 / num2
            print(f"\nResult: {num1} ÷ {num2} = {result}")

    else:
        print("\nInvalid operation selected.")

except ValueError:
    print("\nInvalid input! Please enter numeric values only.")
```
